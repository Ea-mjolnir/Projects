```python
# isolate each word in the text file into a vocabulary dictionary 

import re

def extract_words_from_file(file_path):
    words = []
    with open(file_path, 'r', encoding="utf8") as file:
        file_data = file.read().lower()
        words = re.findall(r'\w+', file_data)
    return set(words)

# Usage example
file_path = '/home/mjolnir/Desktop/DeepL/Projects /Auto_correct_features Using NLP/final.txt'
vocabulary = extract_words_from_file(file_path)

```

```python
import matplotlib.pyplot as plt
from collections import Counter

def count_word_frequency(vocabulary):
    word_counts = {word: 0 for word in vocabulary}
    for word in vocabulary:
        word_counts[word] += 1
    return word_counts

word_freq = count_word_frequency(vocabulary)

# def plot_word_frequency(word_frequency):
#     words = list(word_frequency.keys())
#     frequencies = list(word_frequency.values())
#     plt.bar(words, frequencies)
#     plt.xlabel('Word')
#     plt.ylabel('Frequency')
#     plt.title('Word Frequency')
#     plt.show()
    
# freq=plot_word_frequency(word_freq)


def calculate_word_probabilities(word_list):
   
    word_counts = Counter(word_list)
    # Calculate the total number of words
    total_words = sum(word_counts.values())
    # Calculate the probability of each word
    word_probabilities = {word: count / total_words for word, count in word_counts.items()}

    return word_probabilities

word_probabilities = calculate_word_probabilities(vocabulary)

count=0
# Display the word probabilities
for word, probability in word_probabilities.items():
  print(f"{word}: {probability}")
  count +=1
  if count ==5:
    break
```

```python
import random
import string
from nltk.stem import WordNetLemmatizer


def delete_letter(word):
    variations = set()
    for i in range(len(word)):
        variations.add(word[:i] + word[i+1:])
    return variations

def switch(word):
    variations = set()
    for i in range(len(word) - 1):
        variations.add(word[:i] + word[i+1] + word[i] + word[i+2:])
    return variations

def replace(word):
    variations = set()
    for i in range(len(word)):
        for char in 'abcdefghijklmnopqrstuvwxyz':
            variations.add(word[:i] + char + word[i+1:])
    return variations

def insert(word):
    variations = set()
    for i in range(len(word) + 1):
        for char in 'abcdefghijklmnopqrstuvwxyz':
            variations.add(word[:i] + char + word[i:])
    return variations

def word_variations(word, allow_switches=True):
    variations = set()
    variations.update(delete_letter(word))
    if allow_switches:
        variations.update(switch(word))
    variations.update(replace(word))
    variations.update(insert(word))
    return variations


def word_variation_2(word, allow_switches=True):
    word_variation_2 = set()
    edit_one = word_variations(word, allow_switches=allow_switches)
    for word in edit_one:
        if word:
            edit_two = word_variations(word, allow_switches=allow_switches)
            word_variation_2.update(edit_two)
    return word_variation_2
```
